# 리스트와 반복구조 연습
- 정수로 구성되어 있는 리스트를 파라미터로 입력받아 다음
의 기능을 수행하는 함수 구현

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

print (getIndex(number_list,25))
print (getMax(number_list))
print (getMin(number_list))
print (countGT(number_list,42))
print (sumList(number_list))
swapList(number_list)
print (number_list)
```

###  getIndex(num_list, target)
- 주어진 데이터 target이 어디에 저장되어 있는지 index를 반환

```python
def getIndex(num_list, target):
    return num_list.index(target)
```

### getMax(num_list)
- 리스트에 저장된 숫자 중 가장 큰 숫자 반환

```python
def getMax(num_list) :
    return max(num_list)
```

### getMin(num_list)
- 리스트에 저장된 숫자 중 가장 작은 숫자 반환

```python
def getMin(num_list) :
    return min(num_list)
```

### countGT(num_list, target)
- 리스트에 저장된 숫자 중 입력된 숫자 target보다 큰 수가 몇 개 있는지
구하여 반환

```python
def countGT(num_list, target):
    count=0
    for i in num_list:
        if int(i)>target:
            count+=1
    return count
```

### sumList(nun_list)
- 리스트에 저장된 모든 값 더하여 반환

```python
def sumList(num_list) :
    return sum(num_list)
```

###  swapList(num_list)
- 리스트에 저장된 숫자를 역순으로 저장

```python
def swapList(num_list):
    return num_list.reverse()
```
